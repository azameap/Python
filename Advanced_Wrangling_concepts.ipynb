```python
#lets define a function
def upper_case(column_name):
  return column_name.upper()
```

```python
#after defining a function we make a new column in which we display the upper case charachters
imdb_df['Capital_column'] # lets name the column as capital_column
imdb_df['Capital_column'] = imdb_df['title'] #we want to apply the changes in this title column
imdb_df['Capital_column'] = imdb_df['title'].apply(upper_case)  #here to apply the function to the column we use apply function in pandas
```

```python
import numpy as np
```

```python
def revised_profit(revenue, budget):
  if revenue > 0:
    new_profit = revenue - budget
  else:
    new_profit = np.nan

  return new_profit
```

```python
imdb_df['new_profit']= imdb_df.apply(lambda x:revised_profit(x['revenue'],x['budget']),axis = 1)
imdb_df[['revenue','budget','new_profit']].head()
```

```python
path = '/content/drive/MyDrive/Datasets/imdb_data.csv'
imdb_df = pd.read_csv(path)
imdb_df['Profit_'] = imdb_df['revenue'] - imdb_df['budget']
imdb_df['profit_margin'] = imdb_df.apply(lambda x: (x['Profit_'] / x['revenue']) * 100, axis=1)
imdb_df['is_profitting'] = imdb_df.apply(lambda x: True if x['profit_margin'] >0 else False, axis = 1)
imdb_df[['profit_margin','is_profitting']].head()
```
