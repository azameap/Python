```python
path = '/content/drive/MyDrive/Datasets/customer_information.csv'
cus_df = pd.read_csv(path)
```

```python
cus_df
```

```python
#Identify missing values: In the dataset, the Age column for customer ID 1007 and
#the Phone column for customer ID 1006 are missing values.

missing_phone = cus_df[cus_df['Phone'].isnull()]
missing_phone
```

```python
missing_age = cus_df[cus_df['Age'].isnull()]
missing_age
```

```python
#Fill in missing values: For the Age column of customer ID 1007, the mean value of the Age column
# can be used as a reasonable approximation, which is 32. For the Phone column of customer ID 1006,
#we can either delete the row or try to obtain the missing phone number by contacting the customer
mean_age = round(cus_df['Age'].mean(),0)
cus_df['Age'].fillna(mean_age, inplace = True)
cus_df
```

```python
cus_df.dropna(subset=['Phone'],inplace = True)
```

```python
cus_df
```

```python
#Inconsistent formats: The Phone column has inconsistent formats -
#some are formatted with dashes and others are formatted with parentheses or periods. To standardize the format,
#we can remove all non-numeric characters and keep only the digits.
cus_df['Phone'] = cus_df['Phone'].str.replace(r'\D','',regex = True)
cus_df
```

```python
#Standardize data types:
#The Phone column should be stored as a string instead of a numeric value to preserve the formatting.
#The Age column can be stored as an integer.
cus_df['Phone'] =cus_df['Phone'].astype(str)
cus_df
cus_df['Age'] = cus_df['Age'].astype(int)
cus_df
```
