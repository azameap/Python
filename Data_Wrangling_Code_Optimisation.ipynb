```python
def task(name):
  print(f'Starting task {name}')
  time.sleep(2)
  print(f'Ending task {name}')

if __name__ == '__main__':
  threads = [threading.Thread(target = task, args = (i,)) for i in range(5)]
  for i in range(len(threads)):
    threads[i].start()
  for i in range(len(threads)):
    threads[i].join()

print('All completed')
```

```python
# create a Memory object to cache function calls
mem = Memory(location='cache')

# define a function to calculate the mean of a list of numbers
@mem.cache
def avg(num):
  print('Calculating...')
  return sum(num)/len(num)
```

```python
#Suppose you have a function that takes a long time to run and returns a dictionary.
#Can you use Joblib to cache the results of this function so that subsequent calls to the
#function with the same arguments return the cached value without running the function again?
```

```python
from joblib import Memory
mem = Memory(location = 'cache')
@mem.cache
def create_dict(key,value):
  print('Starting program...')
  my_dict = {key:value}
  return my_dict

```

```python
create_dict("Azam",20)
```

```python
create_dict("Azam",20)
```

```python
create_dict("Azam",20)
```
